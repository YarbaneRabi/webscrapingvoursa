```python
from bs4 import BeautifulSoup
import requests



pages = [requests.get('https://www.voursa.com/Index.cfm?PN='+str(j)+'&gct=1&sct=11&gv=13') for j in range(2,27)]
liens=[]
for page in pages:

  soup = BeautifulSoup(page.text, 'html.parser')
  img_tags = soup.find_all('img')
  img_urls = [img['src'] for img in img_tags if 'src' in img.attrs]

  liens=liens+img_urls
file = open('/content/liens.txt','w')

for i in liens :
  i=i.replace("smph","bgph")
  i=i.replace("psp","pbp")
  if i.endswith('.jpg') and i.startswith('/produitsedr'):

    file.write('https://www.voursa.com'+i+'\n')
    for j in range(2,7):
        file.write('https://www.voursa.com'+i.replace("pbp",str(j))+'\n')



file.close()
```

```python
import requests


file = open('/content/liens.txt','r')
links=[link.strip() for link in file.readlines()]
print(links)
i=1
for link in links:
   with open('/content/drive/MyDrive/train/images2.0/pic_'+str(i)+'.jpg', 'wb') as handle:
     response = requests.get(link, stream=True)

     if not response.ok:
         print(response)
     else:
       for block in response.iter_content(1024):
           if not block:
              break

           handle.write(block)
       i+=1
#webscrapped more than 3000 images from voursa after manually cheking every single one ,only 130 were usable
```
